```python
import matplotlib.pyplot as plt
import scipy.stats as st
import pandas as pd
import numpy as np
```

## Rigid receptor, fully substituted ligand

```python
df = pd.read_csv('../out/output_energies.csv')
y = pd.read_csv('../data/test.csv')['G']
df['y'] = y[:len(df)]
df['y'] = y[:len(df)] 
df = df[(np.abs(st.zscore(df)) < 3).all(axis=1)]
df2 = df.drop(['y'], axis=1)
avg_energy = df2.mean(axis=1)
best_energy = df['0']
y = list(df['y'])
```

```python
r, p = st.pearsonr(best_energy, y)
plt.scatter(best_energy, y)
plt.figtext(0.13, -0.05, f"Correlation: {r}\nP-value:        {p}\nR squared:    {r**2}")
plt.title("Best docking pose binding energy vs. Paulson affinity value: rigid, pdb=2df3")
plt.show()
```

```python
r, p = st.pearsonr(avg_energy, y)
plt.scatter(avg_energy, y)
plt.figtext(0.13, -0.05, f"Correlation: {r}\nP-value:        {p}\nR squared:    {r**2}")
plt.title("Avg (n=20) docking pose binding energy vs. Paulson affinity value: rigid, pdb=2df3")
plt.show()
```
