```python
import matplotlib.pyplot as plt
import scipy.stats as st
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../out/output_energies.csv')
y = pd.read_csv('../data/test.csv')['G']
best_energy = df['0']
avg_energy = df.mean(axis=1)
```

```python
plt.scatter(best_energy, y)
plt.show()
```

```python
r, p = st.pearsonr(avg_energy, y)

plt.scatter(avg_energy, y)

plt.figtext(0.13, 0.9, f"Correlation: {r}\nP-value:        {p}\nR squared:    {r**2}")

plt.show()
```
